# 1. Overfitting

```python
# 1. Generate Data
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

X, y = make_moons(n_samples=300, noise=0.3, random_state=42)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Visualisasi data
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.show()

# 2. Fit Model & Deteksi Overfitting
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

clf = DecisionTreeClassifier(max_depth=None)  # Pohon yang mendalam
clf.fit(X_train, y_train)

train_accuracy = accuracy_score(y_train, clf.predict(X_train))
test_accuracy = accuracy_score(y_test, clf.predict(X_test))

print(f"Akurasi Pelatihan: {train_accuracy:.4f}")
print(f"Akurasi Pengujian: {test_accuracy:.4f}")
```

```python
# 3. Mengatasi Overfitting
clf_fixed = DecisionTreeClassifier(max_depth=4)  # Batasi kedalaman pohon
clf_fixed.fit(X_train, y_train)

train_accuracy_fixed = accuracy_score(y_train, clf_fixed.predict(X_train))
test_accuracy_fixed = accuracy_score(y_test, clf_fixed.predict(X_test))

print(f"Akurasi Pelatihan (Dengan Batasan): {train_accuracy_fixed:.4f}")
print(f"Akurasi Pengujian (Dengan Batasan): {test_accuracy_fixed:.4f}")
```

Dalam contoh di atas:

- Model pertama akan cenderung overfit karena kita tidak membatasi kedalaman pohon.
- Model kedua memiliki kedalaman yang dibatasi, yang membantu mengurangi overfitting.

Ini hanya salah satu contoh pendekatan untuk mendeteksi dan mengatasi overfitting. Dalam praktiknya, Anda mungkin perlu mempertimbangkan teknik lain seperti validasi silang, regularisasi, atau bahkan mengumpulkan lebih banyak data.

# 2. Underfitting

```python
# 1. Generate Data
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
X1, y1 = make_moons(n_samples=300, noise=0.3, random_state=42)
X_train1, X_test1, y_train1, y_test1 = train_test_split(X1, y1, test_size=0.3, random_state=42)
```

```python
# Visualisasi data
plt.scatter(X1[:, 0], X1[:, 1], c=y1)
plt.show()
```

```python
# 2. Fit Model & Deteksi Underfitting
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

clf = DecisionTreeClassifier(max_depth=1)  # Pohon yang sangat dangkal
clf.fit(X_train1, y_train1)

train_accuracy = accuracy_score(y_train1, clf.predict(X_train1))
test_accuracy = accuracy_score(y_test1, clf.predict(X_test1))

print(f"Akurasi Pelatihan: {train_accuracy:.4f}")
print(f"Akurasi Pengujian: {test_accuracy:.4f}")
```

```python
# 3. Mengatasi Underfitting
clf_improved = DecisionTreeClassifier(max_depth=4)  # Tingkatkan kedalaman pohon
clf_improved.fit(X_train1, y_train1)

train_accuracy_improved = accuracy_score(y_train1, clf_improved.predict(X_train1))
test_accuracy_improved = accuracy_score(y_test1, clf_improved.predict(X_test1))

print(f"Akurasi Pelatihan (Diperbaiki): {train_accuracy_improved:.4f}")
print(f"Akurasi Pengujian (Diperbaiki): {test_accuracy_improved:.4f}")
```

Dalam contoh di atas:

- Model pertama akan cenderung underfit karena kita membatasi kedalamannya dengan ekstrim.
- Model kedua memiliki kedalaman yang lebih dalam, yang membantu mengurangi underfitting.

Untuk benar-benar mengatasi underfitting, Anda mungkin juga perlu mempertimbangkan untuk:

- Menggunakan model yang lebih kompleks atau menambahkan fitur-fitur tambahan.
- Menggunakan ensemble learning seperti Random Forest atau Gradient Boosting.
- Menerapkan teknik feature engineering untuk mengekstrak informasi yang lebih bermanfaat dari data.
