# 1. Pembentukan Fold

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import KFold, cross_val_score
import pandas as pd
```

```python
data = load_iris()
X = data.data    # Fitur
y = data.target  # Label
```

```python
# fungsi untuk memecah data ke beberapa bagian (fold_
kfold = KFold(n_splits=5, shuffle=True, random_state=42)
```

```python
# membentuk fold
ls_fold = []
for fold, (train_idx, test_idx) in enumerate(kfold.split(X), 1):
    X_train, X_test = X[train_idx], X[test_idx]
    y_train, y_test = y[train_idx], y[test_idx]
    ls_fold.append([fold,X_train,X_test,y_train,y_test])
```

```python
df_fold = pd.DataFrame(ls_fold,columns=['fold','X_train','X_test','y_train','y_test'])
```

# 2. Fit model tiap fold

```python
from sklearn.linear_model import LogisticRegression
import numpy as np
```

```python
model = LogisticRegression(max_iter=200)
```

```python
scores = []
for fold, (train_idx, test_idx) in enumerate(kfold.split(X), 1):
    X_train, X_test = X[train_idx], X[test_idx]
    y_train, y_test = y[train_idx], y[test_idx]
    model.fit(X_train, y_train)
    score = model.score(X_test, y_test)
    scores.append(score)
    print(f"Fold {fold}: {score:.4f}")
```

```python
print(f"Rata-rata: {np.mean(scores):.4f} (±{np.std(scores):.4f})")
```

# 3. CV secara otomatis

```python
scores_auto = cross_val_score(model, X, y, cv=kfold)
print(f"\nOtomatis: {scores_auto.mean():.4f} (±{scores_auto.std():.4f})")
```

```python
# MODEL FINAL
print("\n=== MELATIH MODEL FINAL (di SELURUH data) ===")
final_model = LogisticRegression(max_iter=200)
final_model.fit(X, y)  # <--- INI YANG DIPAKAI!
```

# 4. Model Final
- Dari 5 fold, tidak ada model yang dipilih.
- Model yang dilatih di setiap fold tidak digunakan lagi. Mereka hanya “alat uji coba” untuk menghitung performa rata-rata.
- Model final = model baru yang dilatih ulang di 100% data menggunakan parameter/hyperparameter terbaik hasil dari CV.
- Hyperparameter = nilai yang kita tentukan sebelum pelatihan dimulai, seperti learning_rate, n_estimators, max_depth, C, kernel, dll. Hyperparameter ini tidak dipelajari dari data — tapi dicari melalui eksperimen.
- Parameter = nilai yang dipelajari oleh model dari data (misalnya koefisien regresi di Linear Regression, bobot dan bias di Neural Network).
- Cara Praktis Mengambil Hyperparameter Terbaik bisa menggunakan **GridSearchCV**

```python
from sklearn.model_selection import cross_val_score

# CV cepat
cv_scores = cross_val_score(LogisticRegression(max_iter=200), X, y, cv=5)
print(f"CV: {cv_scores.mean():.3f}")

# Final model
final_model = LogisticRegression().fit(X, y)
```
